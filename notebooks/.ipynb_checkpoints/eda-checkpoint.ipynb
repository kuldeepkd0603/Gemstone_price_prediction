This dataset contains detailed information about **cubic zirconia gemstones**, with the goal of predicting their **market price** based on physical characteristics and quality grades. Each row represents an individual gemstone, described by both numerical and categorical attributes.

The purpose of this **Exploratory Data Analysis (EDA)** is to:

-  **Understand the structure and distribution** of the dataset.
-  **Detect and handle missing values**, outliers, or inconsistent data.
-  **Explore relationships** between input features and the target variable (Price).
-  **Visualize distributions, correlations**, and category-wise variations to uncover patterns.
-  **Prepare the dataset** for machine learning modeling by identifying key features and transformations needed.

## **3. Data Cleaning**

Data cleaning is the process of removing errors, inconsistencies, and missing values to improve data quality for accurate analysis

### 3.1 Determine  Missing Values

Determining null values involves identifying missing or empty data points in a dataset to handle them appropriately for accurate analysis

```python
data.isnull().sum()
```

we have find null values in one columns :
* Depth: **697**

### 3.2 Handling Missing Values

```python
data['depth'].unique()
```

```python
def impute_depth(row, data):
    if pd.isnull(row['depth']):
        similar_rows = data[
            (np.abs(data['x'] - row['x']) < 0.2) &
            (np.abs(data['y'] - row['y']) < 0.2) &
            (np.abs(data['z'] - row['z']) < 0.1) &
            (data['depth'].notnull())
            ]
            if not similar_rows.empty:
            return similar_rows['depth'].mean()
        else:
            mean_diameter = (row['x'] + row['x']) / 2
            if mean_diameter != 0:
                return (row['z'] / mean_diameter) * 100
    return row['depth']
```

```python
data['depth'] = data.apply(lambda row: impute_depth(row, data), axis=1)
```

This function fills missing depth values by first averaging the depths of diamonds with very similar x, y, and z dimensions, and if none are found, computes the depth percentage directly as (z / mean_diameter) × 100. Otherwise, it simply returns the existing depth

## **4 Univariate Analysis**

### **4.1 Numerical Columns**  

```python
num_cols = ['carat', 'depth', 'table', 'x', 'y', 'z', 'price']
for col in num_cols:
    plt.figure()
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.savefig(f"numerical_{col}.png")
    plt.show()
```

#### 1. Carat (Weight of the Diamond)
- Most cubic zirconia are light and weigh less than 1 carat.
- Heavier ones exist but are rare.
- **Insight**: Smaller stones are more common and popular.

#### 2. Depth (% of height relative to width)
- Most stones have depth between 61–62%.
- This is considered a standard range for a well-cut cubic zirconia.
- **Insight**: Most stones are properly cut and balanced in height.

#### 3. Table (Top flat surface size as %)
- Table size mostly ranges between 54%–60%.
- These values are standard for good quality cuts.
- **Insight**: Most stones have a nicely proportioned top surface.

#### 4. Price
- Most cubic zirconia are priced under $5,000.
- Very high-priced stones are rare.
- **Insight**: Affordable stones are more commonly available.

#### 5. Dimensions (X, Y, Z)
- Most stones are about 4 to 7 mm in length and width, and 2.5 to 4.5 mm in height.
- Sizes are consistent, meaning the shapes are well-balanced.
- **Insight**: Most cubic zirconia are symmetrical and of standard size.

### 5.2 Categorical Columns

```python
cat_cols = ['cut', 'color', 'clarity']
for col in cat_cols:
    plt.figure()
    sns.countplot(data=data, x=col, order=data[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.savefig(f"categorical_{col}.png")
    plt.show()
```

#### 6. Clarity

- Rare types: IF and I1 (Internally Flawless and heavily flawed).
- **Insight**: Most people buy stones with minor flaws to save money without losing much quality.

#### 7. Color
- Common colors: G, E, F (near colorless).
- Least common: J (the most tinted).
- **Insight**: Slightly tinted stones are more popular due to better value.

#### 8. Cut
- Popular cuts: Ideal, Premium, Very Good.
- Less common: Good, Fair.
- **Insight**: Buyers prefer cuts that look brilliant and shiny.

## **5. Bivariate Analysis** 

### 5.1 Numerical vs. Numerical

```python
sns.pairplot(data[['carat', 'x', 'y', 'z', 'price']])
plt.suptitle("Pairwise Scatter Plot", y=1.02)
plt.savefig("numerical_numrical.png")
```

#### 1. Carat vs Price
- Larger stones (more carat) are much more expensive.
- Carat has the strongest effect on price (very high correlation).
- **Insight**: Bigger = more expensive.

#### 2. Dimensions (X, Y, Z) vs Price
- All three dimensions grow with carat and price.
- These features are very similar to carat, so they repeat the same info.
- **Insight**: we Don't use all together in models; choose carat or one of the dimensions.

#### 3. Depth and Table vs Price
- Very weak relationship with price.
- These values don’t help much in predicting price.
- **Insight**: You can include them in a model, but they won’t make a big difference.


### 5.1 Categorical vs. Numerical

```python
for cat in cat_cols:
    plt.figure()
    sns.boxplot(data=data, x=cat, y='price')
    plt.title(f'Price vs {cat}')
    plt.xticks(rotation=45)
    plt.savefig(f"numerical_vs_numerical_{cat}.png")
    plt.show()
```

#### 1. Price vs Clarity
- Higher clarity doesn’t always mean higher price.
- Lower clarity stones (like SI1, SI2) sometimes cost more than flawless ones.
- **Insight**: Size can matter more than clarity in price.

#### 2. Price vs Color
- Stones with worse color (like I or J) can be more expensive.
- Likely because bigger stones tend to have lower color grades.
- **Insight**: Size again plays a big role; color alone doesn’t decide price.

#### 3. Price vs Cut
- All cut types have similar prices.
- “Ideal” cut doesn’t always have the highest price.
- **Insight**: Cut quality is important, but it doesn’t guarantee a higher price.

#### **6.3 Correlation Heatmap**

```python
plt.figure(figsize=(8, 6))  
sns.heatmap(data[num_cols].corr(), annot=True, fmt=".2f", cmap='cividis', square=True)
plt.title("Correlation Matrix")
plt.tight_layout() 
plt.savefig("heatmap.png")
plt.show()
```

1. **Carat (diamond weight) is the most important factor**  
   - Bigger diamonds (higher carat) usually cost more.  
   - There's a strong link between size and price, so carat should always be used in the model.

2. **Dimensions (x, y, z) are very similar to carat**  
   - Length, width, and height of the diamond also affect price.  
   - But they are almost the same as carat, so don’t use all of them together to avoid confusion.

3. **Depth and Table don’t affect price much**  
   - These features have very little impact on the price.  
   - You can still include them, but they won’t help much in predicting price.

#### Grouped mean price by cut and color

```python
pivot = data.pivot_table(index='cut', columns='color', values='price', aggfunc='mean')
plt.figure()
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title("Average Price by Cut and Color")
plt.savefig("price_cut_color.png")
plt.show()
```

#### Heatmap of Average Price by Cut and Color
- Lower color grades (like I and J) have higher average prices across all cuts.
- Premium cut with J color has the highest price.
- **Insight**: Again, size (carat) likely drives price more than just cut or color.

```python
data['carat_bin'] = pd.qcut(data['carat'], 4, labels=['Small', 'Medium', 'Large', 'Extra Large'])
plt.figure()
sns.boxplot(data=data, x='carat_bin', y='price', hue='clarity')
plt.title("Price Distribution by Carat Size and Clarity")
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("price_cart_size_clirity.png")
plt.show()
```

#### Box Plot: Price by Carat Size and Clarity
- As carat size increases, the price rises sharply.
- Clarity has less effect unless the diamond is large.
- **Insight**: Bigger stones matter more than clarity for price.

##  Key Insights from EDA  

1. **Carat (Size) is the Most Important Factor** – Bigger stones cost a lot more and heavily influence the price.
2. **Clarity, Color, and Cut Alone Don’t Decide the Price** – Even less perfect stones can be expensive if they’re bigger.
3. **Dimensions (x, y, z) Are Strongly Linked to Carat** – Use only one (preferably carat) to avoid confusion in models.
4. **Depth and Table Don’t Matter Much for Price** – They can be included in models but won’t have much effect.
5. **Price is Not Just About Quality** – Real-world pricing reflects size and customer preferences more than just perfection.
6. **Most Stones are Affordable** – Majority of diamonds in the data are under $5,000 and are of standard size and quality.
7. **Insights Can Help Build a Better Model** – Focusing on carat, and then combining with cut, clarity, and color gives the best results.
